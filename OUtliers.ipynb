# Dealing with outliers

### Sometimes Outliers can mess up your data analysis!

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
incomes = np.random.normal(27000, 15000, 10000)
incomes = np.append(incomes, [1000000000])
```

```python
plt.hist(incomes, 50)
plt.show()
```

This is not very helpful, t has skewed my data significantly

```python
# what is the mean of the incomes
incomes.mean()
```

Code block that determines to check for outliers

```python
def reject_outliers(data):
    u = np.median(data)
    s = np.std(data)
    filtered = [e for e in data if (u - 2 * s < e < u + 2 * s)]
    return filtered
```

```python
filtered = reject_outliers(incomes)
plt.hist(filtered, 50)
plt.show()
```
